### ê¸°ë³¸(Pre-requisite) ì½”ë“œ

```python
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor

training_data = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor()
)

test_data = datasets.FashionMNIST(
    root="data",
    train=False,
    download=True,
    transform=ToTensor()
)

train_dataloader = DataLoader(training_data, batch_size=64)
test_dataloader = DataLoader(test_data, batch_size=64)

class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, 10),
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits

model = NeuralNetwork()
```

### í•˜ì´í¼íŒŒë¼ë¯¸í„° (Hyperparameter)
* ì—í­(epoch) ìˆ˜ - ë°ì´í„°ì…‹ì„ ë°˜ë³µí•˜ëŠ” íšŸìˆ˜
* ë°°ì¹˜ í¬ê¸°(batch size) - ë§¤ê°œë³€ìˆ˜ê°€ ê°±ì‹ ë˜ê¸° ì „ ì‹ ê²½ë§ì„ í†µí•´ ì „íŒŒëœ ë°ì´í„° ìƒ˜í”Œì˜ ìˆ˜
* í•™ìŠµë¥ (learning rate) - ê° ë°°ì¹˜/ì—í­ì—ì„œ ëª¨ë¸ì˜ ë§¤ê°œë³€ìˆ˜ë¥¼ ì¡°ì ˆí•˜ëŠ” ë¹„ìœ¨. ê°’ì´ ì‘ì„ìˆ˜ë¡ í•™ìŠµ ì†ë„ê°€ ëŠë ¤ì§€ê³ , ê°’ì´ í¬ë©´ í•™ìŠµ ì¤‘ ì˜ˆì¸¡í•  ìˆ˜ ì—†ëŠ” ë™ì‘ì´ ë°œìƒí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
learning_rate = 1e-3
batch_size = 64
epcohs = 5
```

### ìµœì í™” ë‹¨ê³„ (Optimization Loop)
ìµœì í™” ë‹¨ê³„ì˜ ê° ë°˜ë³µ(iteration)ì„ ì—í­ì´ë¼ê³  ë¶€ë¦…ë‹ˆë‹¤.

í•˜ë‚˜ì˜ ì—í­ì€ ë‹¤ìŒ ë‘ ë¶€ë¶„ìœ¼ë¡œ êµ¬ì„±ë©ë‹ˆë‹¤:
* í•™ìŠµ ë‹¨ê³„(train loop) - í•™ìŠµìš© ë°ì´í„°ì…‹ì„ ë°˜ë³µ(iterate)í•˜ê³  ìµœì ì˜ ë§¤ê°œë³€ìˆ˜ë¡œ ìˆ˜ë ´í•©ë‹ˆë‹¤.
* ê²€ì¦/í…ŒìŠ¤íŠ¸ ë‹¨ê³„(validation/test loop) - ëª¨ë¸ ì„±ëŠ¥ì´ ê°œì„ ë˜ê³  ìˆëŠ”ì§€ë¥¼ í™•ì¸í•˜ê¸° ìœ„í•´ í…ŒìŠ¤íŠ¸ ë°ì´í„°ì…‹ì„ ë°˜ë³µ(iterate)í•©ë‹ˆë‹¤.

### ì†ì‹¤ í•¨ìˆ˜ (Loss function)
ì†ì‹¤ í•¨ìˆ˜(loss function)ëŠ” íšë“í•œ ê²°ê³¼ì™€ ì‹¤ì œ ê°’ ì‚¬ì´ì˜ í‹€ë¦° ì •ë„(degree of dissimilarity)ë¥¼ ì¸¡ì •í•˜ë©°, í•™ìŠµ ì¤‘ì— ì´ ê°’ì„ ìµœì†Œí™”í•˜ë ¤ê³  í•©ë‹ˆë‹¤.
* íšŒê·€ ë¬¸ì œ(regression task)ì— ì‚¬ìš©í•˜ëŠ” nn.MSELoss(í‰ê·  ì œê³± ì˜¤ì°¨(MSE; Mean Square Error))
* ë¶„ë¥˜(classification)ì— ì‚¬ìš©í•˜ëŠ” nn.NLLLoss (ìŒì˜ ë¡œê·¸ ìš°ë„(Negative Log Likelihood))
* nn.LogSoftmaxì™€ nn.NLLLossë¥¼ í•©ì¹œ nn.CrossEntropyLoss ë“±

```python
#ğŸ“Œ ëª¨ë¸ì˜ ì¶œë ¥ ë¡œì§“(logit)ì„ nn.CrossEntropyLossì— ì „ë‹¬í•˜ì—¬ ë¡œì§“(logit)ì„ ì •ê·œí™”í•˜ê³  ì˜ˆì¸¡ ì˜¤ë¥˜ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
loss_fn = nn.CrossEntropyLoss()  # ì†ì‹¤ í•¨ìˆ˜ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤.
```

### ì˜µí‹°ë§ˆì´ì € (Optimizaer)
ìµœì í™” ì•Œê³ ë¦¬ì¦˜ì€ ì´ ê³¼ì •ì´ ìˆ˜í–‰ë˜ëŠ” ë°©ì‹(ì—¬ê¸°ì—ì„œëŠ” í™•ë¥ ì  ê²½ì‚¬í•˜ê°•ë²•(SGD; Stochastic Gradient Descent))ì„ ì •ì˜í•©ë‹ˆë‹¤. <br>
ëª¨ë“  ìµœì í™” ì ˆì°¨(logic)ëŠ” optimizer ê°ì²´ì— ìº¡ìŠí™”(encapsulate)ë©ë‹ˆë‹¤. <br>
ì—¬ê¸°ì„œëŠ” SGD ì˜µí‹°ë§ˆì´ì €ë¥¼ ì‚¬ìš©í•˜ê³  ìˆìœ¼ë©°, PyTorchì—ëŠ” ADAMì´ë‚˜ RMSPropê³¼ ê°™ì€ ë‹¤ë¥¸ ì¢…ë¥˜ì˜ ëª¨ë¸ê³¼ ë°ì´í„°ì—ì„œ ë” ì˜ ë™ì‘í•˜ëŠ” ë‹¤ì–‘í•œ ì˜µí‹°ë§ˆì´ì €ê°€ ìˆìŠµë‹ˆë‹¤.

í•™ìŠµ ë‹¨ê³„(loop)ì—ì„œ ìµœì í™”ëŠ” ì„¸ë‹¨ê³„ë¡œ ì´ë¤„ì§‘ë‹ˆë‹¤:
* optimizer.zero_grad()ë¥¼ í˜¸ì¶œí•˜ì—¬ ëª¨ë¸ ë§¤ê°œë³€ìˆ˜ì˜ ë³€í™”ë„ë¥¼ ì¬ì„¤ì •í•©ë‹ˆë‹¤. ê¸°ë³¸ì ìœ¼ë¡œ ë³€í™”ë„ëŠ” ë”í•´ì§€ê¸°(add up) ë•Œë¬¸ì— ì¤‘ë³µ ê³„ì‚°ì„ ë§‰ê¸° ìœ„í•´ ë°˜ë³µí•  ë•Œë§ˆë‹¤ ëª…ì‹œì ìœ¼ë¡œ 0ìœ¼ë¡œ ì„¤ì •í•©ë‹ˆë‹¤.
* loss.backwards()ë¥¼ í˜¸ì¶œí•˜ì—¬ ì˜ˆì¸¡ ì†ì‹¤(prediction loss)ì„ ì—­ì „íŒŒí•©ë‹ˆë‹¤. PyTorchëŠ” ê° ë§¤ê°œë³€ìˆ˜ì— ëŒ€í•œ ì†ì‹¤ì˜ ë³€í™”ë„ë¥¼ ì €ì¥í•©ë‹ˆë‹¤.
* ë³€í™”ë„ë¥¼ ê³„ì‚°í•œ ë’¤ì—ëŠ” optimizer.step()ì„ í˜¸ì¶œí•˜ì—¬ ì—­ì „íŒŒ ë‹¨ê³„ì—ì„œ ìˆ˜ì§‘ëœ ë³€í™”ë„ë¡œ ë§¤ê°œë³€ìˆ˜ë¥¼ ì¡°ì •í•©ë‹ˆë‹¤.

```python
#ğŸ“Œ í•™ìŠµí•˜ë ¤ëŠ” ëª¨ë¸ì˜ ë§¤ê°œë³€ìˆ˜ì™€ í•™ìŠµë¥ (learning rate) í•˜ì´í¼íŒŒë¼ë¯¸í„°ë¥¼ ë“±ë¡í•˜ì—¬ ì˜µí‹°ë§ˆì´ì €ë¥¼ ì´ˆê¸°í™”
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

### ì „ì²´ êµ¬í˜„
ìµœì í™” ì½”ë“œë¥¼ ë°˜ë³µí•˜ì—¬ ìˆ˜í–‰í•˜ëŠ” train_loopì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ ì¸¡ì •í•˜ëŠ” test_loopë¥¼ ì •ì˜í•˜ì˜€ìŠµë‹ˆë‹¤.

```python
def train_loop(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    # ëª¨ë¸ì„ í•™ìŠµ(train) ëª¨ë“œë¡œ ì„¤ì •í•©ë‹ˆë‹¤ - ë°°ì¹˜ ì •ê·œí™”(Batch Normalization) ë° ë“œë¡­ì•„ì›ƒ(Dropout) ë ˆì´ì–´ë“¤ì— ì¤‘ìš”í•©ë‹ˆë‹¤.
    # ì´ ì˜ˆì‹œì—ì„œëŠ” ì—†ì–´ë„ ë˜ì§€ë§Œ, ëª¨ë²” ì‚¬ë¡€ë¥¼ ìœ„í•´ ì¶”ê°€í•´ë‘ì—ˆìŠµë‹ˆë‹¤.
    model.train()
    for batch, (X, y) in enumerate(dataloader):
        # ì˜ˆì¸¡(prediction)ê³¼ ì†ì‹¤(loss) ê³„ì‚°
        pred = model(X)
        loss = loss_fn(pred, y)

        # ì—­ì „íŒŒ
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        if batch % 100 == 0:
            loss, current = loss.item(), batch * batch_size + len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")


def test_loop(dataloader, model, loss_fn):
    # ëª¨ë¸ì„ í‰ê°€(eval) ëª¨ë“œë¡œ ì„¤ì •í•©ë‹ˆë‹¤ - ë°°ì¹˜ ì •ê·œí™”(Batch Normalization) ë° ë“œë¡­ì•„ì›ƒ(Dropout) ë ˆì´ì–´ë“¤ì— ì¤‘ìš”í•©ë‹ˆë‹¤.
    # ì´ ì˜ˆì‹œì—ì„œëŠ” ì—†ì–´ë„ ë˜ì§€ë§Œ, ëª¨ë²” ì‚¬ë¡€ë¥¼ ìœ„í•´ ì¶”ê°€í•´ë‘ì—ˆìŠµë‹ˆë‹¤.
    model.eval()
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0

    # torch.no_grad()ë¥¼ ì‚¬ìš©í•˜ì—¬ í…ŒìŠ¤íŠ¸ ì‹œ ë³€í™”ë„(gradient)ë¥¼ ê³„ì‚°í•˜ì§€ ì•Šë„ë¡ í•©ë‹ˆë‹¤.
    # ì´ëŠ” requires_grad=Trueë¡œ ì„¤ì •ëœ í…ì„œë“¤ì˜ ë¶ˆí•„ìš”í•œ ë³€í™”ë„ ì—°ì‚° ë° ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰ ë˜í•œ ì¤„ì—¬ì¤ë‹ˆë‹¤.
    with torch.no_grad():
        for X, y in dataloader:
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()

    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

epochs = 10
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train_loop(train_dataloader, model, loss_fn, optimizer)
    test_loop(test_dataloader, model, loss_fn)
print("Done!")
```
